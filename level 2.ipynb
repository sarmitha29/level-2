```python
import pandas as pd
```

```python
data = pd.read_csv("D:\Dataset.csv")
```

```python
data
```

```python
data['Has Table booking'] = data['Has Table booking'].map({'Yes': True, 'No': False})
data['Has Online delivery'] = data['Has Online delivery'].map({'Yes': True, 'No': False})
# Calculate the percentage of restaurants offering table booking and online delivery
total_restaurants = len(data)
table_booking_percentage = (data['Has Table booking'].sum() / total_restaurants) * 100
online_delivery_percentage = (data['Has Online delivery'].sum() / total_restaurants) * 100

print(f"Percentage of restaurants offering table booking: {table_booking_percentage:.2f}%")
print(f"Percentage of restaurants offering online delivery: {online_delivery_percentage:.2f}%")
```

```python
# Find the most common price range among all restaurants
most_common_price_range = data['Price range'].mode()[0]
print(f"The most common price range among all restaurants is: {most_common_price_range}")
```

```python
# Calculate the average rating for each price range
average_rating_per_price_range = data.groupby('Price range')['Aggregate rating'].mean()
print("Average rating for each price range:")
print(average_rating_per_price_range)
```

```python
# Find the color representing the highest average rating among different price ranges
highest_rating_color = data.groupby('Price range')['Aggregate rating'].mean().idxmax()
print(f"The color representing the highest average rating among different price ranges is: {highest_rating_color}")
```

```python
# Extracting length of restaurant name and address
data['Restaurant Name length'] = data['Restaurant Name'].apply(len)
data['Address length'] = data['Address'].apply(len)
```

```python
# Encoding categorical variables like 'Has Table Booking' and 'Has Online Delivery'
data['has table booking'] = data['Has Table booking'].replace({'Yes': 1, 'No': 0}).astype(int)
data['has online delivery'] = data['Has Online delivery'].replace({'Yes': 1, 'No': 0}).astype(int)
```

```python
# Displaying the modified dataset with new features
print("Modified Dataset with New Features:")
data  # Show the first few rows of the modified dataset
```

```python
# Summary statistics of the new features
print("\nSummary Statistics of New Features:")
print(data[['Restaurant Name length', 'Address length', 'has table booking', 'has online delivery']].describe())
```
